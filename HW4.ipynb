```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

def numb_in_doc(number):
    '''
    Проверят есть ли номер в базе
    '''
    for i in documents:
        if i['number'] == number:
            return True
    return False
    
def get_number():
    '''
    Возвращает данные о владельце исходя из номера его документа
    '''
    numb = input('Ввведите номер документа')
    exist = numb_in_doc(numb)
    if exist == True:
            for i in documents:
                if numb == i['number']:
                    result = 'Владелец документа: '+ i['name']
    else:
        result = 'Документ не найден в базе'
    print(result)

def get_shelf():
    '''
    Возвращает данные о месте хранения исходя из номера его документа
    '''
    numb = input('Ввведите номер документа')
    list_numb = []
    for i in directories.values():
        for ii in i:
            list_numb.append(ii)
    if numb in list_numb:
        for item, values in directories.items():
            if numb in values:
                result = 'Документ хранится на полке: ' + item
    else:
        result = 'Документ не найден в базе'
    print(result)

def get_shelf1(numb):
    '''
    Возвращает данные о месте хранения исходя из номера его документа
    '''
    for item, values in directories.items():
            if numb in values:
                result =  item
    return result
def all_info():
    '''
    Информация о всех документах и месте хранения
    '''
    for i in documents:
        nn = i['number']
        typei = i['type']
        namei = i['name']
        result = get_shelf1(nn)
        info = print(f'#: {nn}, тип: {typei}, владелец: {namei}, полка хренения: {result}')

def adsh():
    '''
    Добавляет новую полку
    '''
    nums = input('Введите номер полки')
    if nums not in directories:    
        directories[nums]  = []
        res = ', '.join(list(directories.keys()))
        print (f'Полка добавлена. Текущий уровень полок: {res}')
    else:
        res = ', '.join(list(directories.keys()))
        print (f'Такая полка уже существует. Текущий уровень полок: {res}')

def dell():
    '''
    Удаляет полку если она не пустая
    '''
    nums = input('Введите номер полки') 
    sheld = []
    list_sh = list(directories.keys())
    for key, value in directories.items():
        if value:
            sheld.append(key)
    if nums in list_sh:    
        if nums not in sheld:
            del(directories[nums])
            print (f'Полка удалена. Текущий перечень полок: {list(directories.keys())}')
        else:
            print(f'На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {list_sh}')
    else:
        print(f'Такой полки не существует. Текущий перечень полок: {list_sh}')


def add_data():
    '''
    Добавляет нового владельца
    '''
    sh_n = input('Введите полку хранения для докмента')
      
    if sh_n in directories: 
        num_d = input('Введите номер докмента')
        type_d = input('Введите тип докмента')
        person = input('Введите владельца докмента')
        documents.append(dict(type=type_d, number=num_d, name=person))
        for key, values in directories.items():
            if sh_n == key:
                directories[key] += [num_d]
                print('Документ добавлен. Текущий список документов:')
                all_info()
    else:
        print('Такой полки не существует. Добавьте полку командой as.\nТекущий список документов:')
        all_info()


def del_pers():
    '''
    Удаляет документ из данных
    '''
    num_d = input('Введите номер документа') 
    for i in documents:
        if num_d == i['number']:
            res = i
            n_index = documents.index(res)        
            del documents[n_index]
            resp = 'Документ удален.\nТекущий список документов: '
        else:
            resp = 'Документ не найден в базе.\nТекущий список документов:'
    return print(resp), all_info()

def move_num():
    '''
    Перемещает документ с полки на полку
    '''
    numb  = input('Введите номер')
    nums = input('Введите номер полки')
    key_list = list(directories.keys())
    numb_list = []
    for i in documents:
        numb_list.append(i['number'])
    if numb in numb_list:
        if nums in key_list:
            for key, value in directories.items():
                if numb in value:
                    value.remove(numb)
            directories[nums] += [numb] 
            print('Документ перемещен.\nТекущий список документов:')
            all_info()
        else:
            strr = ', '.join(key_list)
            print(f'Такой полки не существует. Текущий перечень полок: {strr}.')
    else:
        print('Документ не найден в базе.\nТекущий список документов:')
        all_info()
        
def main():
    '''
    Основная функция
    '''
    inp_dict = {'p' : get_number, 's' : get_shelf, 'l' : all_info,
                'as' : adsh, 'ds' : dell, 'ad' : add_data, 'd' : del_pers, 'm' : move_num,'q' : False}
    y = True
    while y:
        choice = input('Введите команду').lower()
        if choice in inp_dict.keys():
            if choice == 'q':
                y = False
            else:
                inp_dict[choice]()
        else:
            print('Такой команды нет')
            
    
        
main()
    
```
